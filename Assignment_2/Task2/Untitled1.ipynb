```python
import matplotlib.pyplot as plt
```

```python
aann1 = []
with open("aann1_loss.txt","r") as f:
    lines = f.readlines()
    for line in lines:
        c = line.strip().split()
        aann1.append(float(c[len(c)-1]))   

aann2 = []
with open("aann2_loss.txt","r") as f:
    lines = f.readlines()
    for line in lines:
        c = line.strip().split()
        aann2.append(float(c[len(c)-1]))   

aann3 = []
with open("aann3_loss.txt","r") as f:
    lines = f.readlines()
    for line in lines:
        c = line.strip().split()
        aann3.append(float(c[len(c)-1]))  

sae = []
with open("sae_loss.txt","r") as f:
    lines = f.readlines()
    for line in lines:
        c = line.strip().split()
        sae.append(float(c[len(c)-1]))  

    
```

```python
import numpy as np
aann1 = np.array(aann1)
aann2 = np.array(aann2)
aann3 = np.array(aann3)
sae = np.array(sae)
```

```python
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss v/s Epoch for AANN1")
plt.plot(aann1)
plt.savefig("aann1_loss.png")
```
